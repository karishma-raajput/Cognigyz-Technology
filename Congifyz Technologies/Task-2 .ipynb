```python
data['Restaurant Name'].sample(20,random_state=194)
```

```python
restaurantXcuisines=pd.crosstab(data['Restaurant Name'],data['Cuisines'])
restaurantXcuisines
```

```python
print(jaccard_score(restaurantXcuisines.loc['Olive Bistro'].values,restaurantXcuisines.loc['Rose Cafe'].values, average='macro'))
```

```python
jaccardDist = pdist(restaurantXcuisines.values,metric='jaccard')
jaccardMatrix = squareform(jaccardDist)
jaccardSim=1 - jaccardMatrix
dfJaccard = pd.DataFrame(
    jaccardSim,
    index= restaurantXcuisines.index,
    columns= restaurantXcuisines.index
)
dfJaccard
```

```python
new_data['Restaurant Name'].sample(20)
```

```python
restaurant="Ooma"
sim = dfJaccard.loc[restaurant].sort_values(ascending=False)
sim = pd.DataFrame({'Restaurant Name': sim.index, 'simScore': sim.values})
sim = sim[(sim['Restaurant Name']!= restaurant) & (sim['simScore']>=0.7)].head(5)
Restaurant_Rec=pd.merge(sim,new_data[['Restaurant Name','Aggregate rating']],how='inner',on='Restaurant Name')
final_Restaurant_Rec = Restaurant_Rec.sort_values('Aggregate rating',ascending=False).drop_duplicates('Restaurant Name',keep='first')
final_Restaurant_Rec
```
